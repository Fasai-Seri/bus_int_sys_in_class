```python
url = "https://www.visualcapitalist.com/electric-vehicle-sales-by-model-2023/"
dfs = pd.read_html(url)
dfs
```

```python
import requests
import io
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
r = requests.get(url, headers={'user-agent': 'Mozilla'})
if r.ok:
    # r.text
    dfs = pd.read_html(io.StringIO(r.text)) # convert text into file
    print(len(dfs))
```

```python
dfs[0]
```

```python
dfs[1]
```

```python
df = dfs[0]
df
```

```python
def get_country(s):
    flag, country_name = s.split(' ')
    return country_name

# get_country("🇨🇳 China")
```

```python
df['Country'] = df['Country'].map(get_country) # functional programming
df
```

```python
# format ticker
import matplotlib.ticker as ticker
g=sns.barplot(data=df, y="Model", x="Vehicles Sold (Jan-Aug 2023)");
g.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f"{x/1e6:,.1f}m")) 
```
